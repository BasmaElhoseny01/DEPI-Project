## English Sentiment

## cardiffnlp/twitter-roberta-base-sentiment

https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Loading a Pre-Trained Model from HuggingFace Hub
model_name = "cardiffnlp/twitter-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create Pipeline
sentiment_pipe = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer,return_all_scores=True,device="cuda")
```

```python
# Running the Application
result = sentiment_pipe("I love using HuggingFace models for NLP tasks!")
print(f"Classification: {result}")
```

```python
sentiment_samples = [
    "I love this!",               # Positive
    "This is the worst!",         # Negative
    "I'm not sure how I feel.",   # Neutral
    "What a fantastic experience!", # Positive
    "I absolutely hate it.",      # Negative
    "It's okay, I guess.",        # Neutral
    "This makes me so happy.",    # Positive
    "I'm really disappointed.",   # Negative
    "Not bad, but could be better.", # Neutral
    "I'm thrilled with the result!" # Positive
]
sentiment_pipe(sentiment_samples)
```

## Arabic Sentiment

### CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment

https://huggingface.co/CAMeL-Lab/bert-base-arabic-camelbert-da-*sentiment*

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_name = "CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment"

arabic_tokenizer = AutoTokenizer.from_pretrained(model_name)
arabic_model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create Pipeline
arabic_sentiment_pipe = pipeline('sentiment-analysis', model=arabic_model, tokenizer=arabic_tokenizer,device ="cuda")
```

```python
sentences = ['أنا بخير', 'أنا لست بخير','أنا أحب الأطفال',"أنا لا أحب الأطفال","أنا أكره البيتزا"]
arabic_sentiment_pipe(sentences)
```

```python
arabic_rsentiment_samples = [
    "أنا بخير",
    "أنا لست بخير",
    "أنا أحب الأطفال",
    "أنا لا أحب الأطفال",
    "أنا أكره البيتزا",
    "الجو جميل اليوم",
    "لا أشعر بالسعادة",
    "هذا أفضل يوم في حياتي",
    "هذا اليوم كان سيئاً",
    "أنا متحمس جداً للمستقبل"
]

arabic_sentiment_pipe(arabic_rsentiment_samples)
```
